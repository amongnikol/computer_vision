# Optical Character Recognition with EasyOCR

```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
IMAGE_PATH = 'struk.jpg'
```

#Print All Text

```python
reader = easyocr.Reader(['id'])
result = reader.readtext(IMAGE_PATH)
result
```

# Text Marking

```python
#Koordinat
result[0][0]
```

```python
#Pojok kiri atas dan kanan bawah
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
print(top_left)
```

# All Text Marking

```python
img = cv2.imread(IMAGE_PATH)

for i in range(len(result)):
  top_left = tuple(result[i][0][0])
  bottom_right = tuple(result[i][0][2])
  img = cv2.rectangle(img,top_left,bottom_right,(255,0,0),1)
```

```python
plt.imshow(img)
plt.show()
```
